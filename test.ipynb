```python
from pwdlib import PasswordHash

pwd = PasswordHash.recommended()
```

```python
a = pwd.hash("hello!")
a
```

```python
pwd.verify_and_update("hello!", a)
```

```python
from pwdlib.hashers.bcrypt import BcryptHasher

bpwd = BcryptHasher()
a = bpwd.hash("hello!")
a
```

```python
pwd.verify_and_update("hello!", a)
```

```python
ah = PasswordHash(hashers=[pwd, bpwd])
ah.verify_and_update("hello!", a)
```

```python
from pwdlib.hashers.argon2 import Argon2Hasher
ah = PasswordHash(hashers=[Argon2Hasher(), BcryptHasher()])
ah.verify_and_update("hello!", a)
```
